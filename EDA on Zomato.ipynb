# Reading CSV

```python
df=pd.read_csv("zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url', 'address','phone','reviews_list','menu_item','dish_liked'],axis= 1)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df["rate"].unique()
```

# Removing null value from rate column i.e("new","-", "/5")

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value= str(value).split("/")
        value=value[0]
        return float(value)
df["rate"]=df["rate"].apply(handlerate)
df["rate"].head()
```

```python
df.rate.isnull().sum()
```

# Filling Null Values with mean in Rate Column

```python
df["rate"].fillna(df["rate"].mean(), inplace=True)
df["rate"].isnull().sum()
```

```python
df.info()
```

# Dropping null values

```python
df.dropna(inplace=True)
df.info()
```

```python
df.head()
```

```python
df.rename(columns={"approx_cost(for two people":"Cost2Plates","listed_in(type)":"Type"}, inplace=True)
```

```python
df.head()
```

```python
df=df.drop(["listed_in(city)"],axis=1)
```

```python
df["approx_cost(for two people)"].unique()
```

```python
def handelcomma(value):
    value = str(value)
    if ',' in value:
        value=value.replace(',','')
        return int (value)
    else:
        return int(value)
df["approx_cost(for two people)"]=df["approx_cost(for two people)"].apply(handelcomma)
df["approx_cost(for two people)"].unique()
```

```python
df.head()
```

# Cleaning Rest Type

```python
df["rest_type"].value_counts()
```

```python
df["rest_type"].value_counts()
```

```python
rest_type=df["rest_type"].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessthan1000=rest_type[rest_type<1000]
rest_type_lessthan1000
```

# Making Rest Types less than 1000 in frequency as other

```python
def handle_rest_type(value):
    if(value in rest_type_lessthan1000):
        return "others"
    else:
        return value
df["rest_type"]=df["rest_type"].apply(handle_rest_type)
df["rest_type"].value_counts()
```

```python
df.head()
```

```python
df["location"].value_counts()
```

```python
location=df["location"].value_counts(ascending=False)
location_lessthan300=location[location<300]

def handle_location(value):
    if (value in location_lessthan300):
        return"others"
    else:
        return value
    
df["location"]=df["location"].apply(handle_location)
df["location"].value_counts()
```

```python
df.head()
```

# Cleaning Cusines Column

```python
df["cuisines"].value_counts()
```

```python
cuisines=df["cuisines"].value_counts(ascending=False)
cuisines_lessthan100=cuisines[cuisines<100]

def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return "others"
    else:
        return value
df["cuisines"]=df["cuisines"].apply(handle_cuisines)
df["cuisines"].value_counts()
```

```python
df.head()
```

```python
df["Type"].value_counts()
```
